```python
import pandas as pd
import numpy as np
import sklearn.datasets
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
bc_data = sklearn.datasets.load_breast_cancer()
print(bc_data)
```

```python
BC_data = pd.DataFrame(data=bc_data.data,columns=bc_data.feature_names)
BC_data.head()
```

```python
BC_data['Label'] = bc_data.target
BC_data.head()
```

1 -- Benign

0 -- Malignant

```python
BC_data.groupby('Label').mean()
```

```python
X = BC_data.drop(columns='Label',axis=1)
Y = BC_data['Label']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_tr = scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_tr,Y,test_size=0.1,stratify=Y,random_state=2)
```

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_split=0.1,epochs=10)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.legend(['training_data','validation_data','Loss','val_Loss'])
plt.xlabel('Epoch')
plt.ylabel('Accuracy and loss')

plt.show()
```

```python
h = model.evaluate(x_test,y_test)
print(h)
```

```python
ypre = model.predict(x_test)
print(ypre[0])
```
