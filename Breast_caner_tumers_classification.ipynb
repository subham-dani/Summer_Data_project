```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
BC_data = pd.read_csv('/content/sample_data/breast_cancer_data.csv')
BC_data.head()
```

```python
BC_data.info()
```

```python
BC_data['diagnosis'].value_counts()
```

```python
NBC_data = BC_data.drop(columns=['Unnamed: 32','id'],axis=1)
NBC_data.head()
```

```python
X = NBC_data.drop(columns=['diagnosis'],axis=1)
Y = NBC_data['diagnosis']
```

```python
scaler = StandardScaler()
X_tr = scaler.fit_transform(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_tr,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
ypre = model.predict(x_train)
ypre_1 = model.predict(x_test)
acc = accuracy_score(y_train,ypre)
acc_1 = accuracy_score(y_test,ypre_1)
print(acc, acc_1)
```
